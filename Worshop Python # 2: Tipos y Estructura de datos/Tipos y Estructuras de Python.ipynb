## Números Complejos

> son aquellos que tienen dos partes:
- Una parte real, como por ejemplo 3 o 7.5.
- Y otra imaginaria, como 5j o -7j.

> parte_real + parte_imaginaria*j

```python
a = 5 + 5j
b = 1.3 - 7j
```

### ¿Para qué sirven los números complejos?
>Los números complejos son muy utilizados en las telecomunicaciones y electrónica, ya que son muy útiles para describir las ondas electromagnéticas y la corriente eléctrica.

>También son usados en diferentes dominios de las matemáticas y en física, donde destaca su uso en la mecánica cuántica.

>Un número complejo puede representarse en un plano, donde el eje x representa la parte real y el eje y la imaginaria. Es decir, se puede ver a un número complejo 5+5i como un punto de coordenadas.

image.png

>Una vez representado en esta gráfica, cualquier número complejo formará un ángulo con el eje x. Todo número complejo tiene también un módulo, que es la distancia que une el punto del origen de coordenadas 0+0i.

```python
c = 3 + 5j
print(c)       #(3+5j)
print(type(c)) #<class 'complex'>
```

```python
# Parte real e imaginaria
c = 3 + 5j
print(c.real) #3.0
print(c.imag) #5.0
```

### Operaciones con números complejos


```python
# Suma de complejos
a = 1 + 3j
b = 4 + 1j
print(a+b) #(5+4j)
```

```python
# Resta de complejos
a = 1 + 3j
b = 4 + 1j
print(a-b) #(-3+2j)
```

```python
# Multiplicación de complejos
a = 1 + 3j
b = 4 + 1j
print(a*b) #(1+13j)
```

```python
# División de complejos
a = 1 + 3j
b = 4 + 1j
print(a/b) #(0.41+0.64j)
```

## Set
>Los set en Python son un tipo que permite almacenar varios elementos y acceder a ellos de una forma muy similar a las listas pero con ciertas diferencias:

- Los elementos de un set son único, lo que significa que no puede haber elementos duplicados.
- Los set son desordenados, lo que significa que no mantienen el orden de cuando son declarados.
- Sus elementos deben ser inmutables

```python
herramientas = set(["Python", "Airflow", "Snowflake", "SQL", "Databricks"])
print(herramientas)
print(type(herramientas))
```

```python
herramientas = {"Python", "Airflow", "Snowflake", "SQL", "Databricks"}
print(herramientas)
print(type(herramientas))
```

```python
# funcion len

x = set([1, 2, 3, 3, 4, 4, 5])
print(len(x))
```

###Métodos

```python
# El método add() permite añadir un elemento al set.

a = set([1, 2, 3])
a.add(4)
print(a)
```

```python
# El método remove() elimina el elemento que se pasa como parámetro.

a = set([1, 2, 3, 4])
a.remove(4)
print(a)
```

```python
# El método pop() elimina un elemento aleatorio del set.

a = set([1, 2, 3, 4])
a.pop()
print(a)
```

```python
# El método clear() elimina todos los elementos de set.

a = set([1, 2, 3, 4])
a.clear()
print(a)
```

## Diccionario
>Un diccionario en Python es una colección de elementos, donde cada uno tiene una llave key y un valor value. Los diccionarios se pueden crear con paréntesis {} separando con una coma cada par key: value. 

```python
#Diccionario
d1 = {
  "Nombre": "Miguel",
  "Edad": 27,
  "Documento": 1003882
}
print(d1)
```

```python
d2 = dict([
      ('Nombre', 'Miguel'),
      ('Edad', 27),
      ('Documento', 1003882),
])
print(d2)
```

```python
# Acceder a elemtos
print(d1['Nombre'])
print(d1.get('Nombre'))
```

```python
# Modificar elementos

d1['Nombre'] = "Carlos"
print(d1)
```

```python
# Agregar información

d1['Profession'] = "Developer"
print(d1)
```

```python
# Iterar diccionario
# Imprime los key del diccionario

for i in d1:
    print(i)
```

```python
# Impre los value del diccionario

for x in d1:
    print(d1[x])
```

```python
# Imprime los key y value del diccionario

for x, y in d1.items():
    print(x, y)
```

### Métodos diccionarios Python

```python
# El método clear() elimina todo el contenido del diccionario.

d1 = {
  "Nombre": "Carlos",
  "Edad": 27,
  "Documento": 1003882
}
d1.clear()
print(d1)
```

```python
# El método items() devuelve una lista con los keys y values del diccionario.

d1 = {
  "Nombre": "Carlos",
  "Edad": 27,
  "Documento": 1003882
}
it = d1.items()
print(it) 
print(list(it))
print(list(it)[0][0])
```

### ChainMap
>Es una clase similar a un diccionario que puede hacer una vista única de múltiples asignaciones. Básicamente, devuelve una lista de varios otros diccionarios. Suponga que tiene dos diccionarios con varios pares clave-valor, en este caso ChainMap hará una sola lista con ambos diccionarios en ella.

```python
from collections import ChainMap 
```

```python
juguetes = {'muñecas': 30, 'carros': 25}
computadoras = {'Mac': 100, 'HP': 150}
ropa = {'camisas': 20, 'faldas': 20}
inventario = ChainMap(juguetes, computadoras, ropa)
print(inventario)
```

```python
consolas = {'PS5': 5, 'XBOX': 6}
inventario2 = inventario.new_child(consolas)
print(inventario2)

```

### Counter

### OrderedDict
>Es una subclase de diccionario que recuerda el orden en que se agregaron las entradas. Básicamente, incluso si cambia el valor de la clave, la posición no cambiará debido al orden en que se insertó en el diccionario.

```python
 from collections import OrderedDict
```

```python
numero = OrderedDict()
numero["one"] = 1
numero["two"] = 2
numero["three"] = 3
print(numero)
```

### defaultdict
> Es una subclase de diccionario que llama a una función de fábrica para proporcionar valores faltantes. En general, no arroja ningún error cuando se llama a un valor de clave faltante en un diccionario.

```python

from collections import defaultdict 
```

```python
# Acceso al valor de una clave que falta desde un diccionario de Python
data = {'Nombre': 'Jose', 'Ciudad': 'Lima', 'Edad': 33}
print(data['Hobbies'])
```

```python
# Uso de .setdefault() para establecer un valor predeterminado para una clave que falta
data = {'Nombre': 'Jose', 'Ciudad': 'Lima', 'Edad': 33}
data.setdefault('Hobbies', None)

print(data['Hobbies'])
```
