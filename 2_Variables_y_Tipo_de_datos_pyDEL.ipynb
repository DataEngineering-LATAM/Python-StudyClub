```python
x=7

x+=2	#x=x+2 = 9
print(x)

x-=2	#x=x-2 = 7
print(x)

x*=2	#x=x*2 = 14
print(x)
```

```python
#Varias operaciones en una sola l√≠nea
a=7; b=2

print("Operadores de asignaci√≥n")
x=a; x+=b;  print("x+=", x)  # 9
x=a; x-=b;  print("x-=", x)  # 5
x=a; x*=b;  print("x*=", x)  # 14
```

```python
# Checking data types

print(type(10))                  # Int
print(type(3.14))                # Float
print(type(1 + 3j))              # Complex
print(type('Data Engineering Latam'))          # String
print(type([1, 2, 3]))           # List
print(type({'comunidad':'Data Engineering Latam'})) # Dictionary
print(type({9.8, 3.14, 2.7}))    # Set
print(type((9.8, 3.14, 2.7)))    # Tuple
```

```python
#Listas 
list1=[1,4,"Data",6,"college"]
list2=[]  # creates an empty list
list3=list((1,2,3))

print(list1)
print(list2)
print(list3)
```

```python
#tuplas

tuple1=(1,2,"college",9)
tuple2=() # creates an empty tuple
tuple3=tuple((1,3,5,9,"hello"))

print(tuple1)
print(tuple2)
print(tuple3)
```

```python
#Set  / Conjuntos 
set1={1,2,3,4,5,"hello","tup"}
set2={(1,8,"python",7)}

print(set1)
print(set2)
```

```python
#Diccionarios 
dict1={"key1":"value1","key2":"value2"}
dict2={}   # empty dictionary
dict3=dict({1:"apple",2:"cherry",3:"strawberry"})

print(dict1)
print(dict2)
print(dict3)
```
